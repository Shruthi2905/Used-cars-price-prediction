```python
processed=processed.drop(['Price'],axis=1)
processed.head()
```

```python
processed.describe(include='all')
```

```python
data_with_dummies=pd.get_dummies(processed,drop_first=True)
data_with_dummies.head()
```

```python
data_with_dummies.describe(include='all')
```

```python
data_with_dummies.columns.values
```

```python
cols=['logprice', 'Mileage', 'EngineV', 'Year', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed=data_with_dummies[cols]
data_preprocessed.head()
```

linear regression

```python
target=data_preprocessed['logprice']
inputs=data_preprocessed.iloc[:,1:]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(inputs)
```

```python
scaled_inputs=scaler.transform(inputs)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,target,test_size=0.2,random_state=10)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
yhat=reg.predict(x_train)
plt.scatter(y_train,yhat)
plt.xlabel('ytrain')
plt.ylabel('yhat')
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.distplot(y_train-yhat)
plt.title("residual check")
```

```python
reg.score(x_train,y_train)
```

```python
reg_summary=pd.DataFrame(inputs.columns.values,columns=['Features'])
reg_summary['coef']=reg.coef_
reg_summary
```

```python
df_pf=pd.DataFrame(np.exp(yhat_test),columns=['predictions'])
```

```python
df_pf
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
df_pf['targets']=np.exp(y_test)
df_pf
```

```python
df_pf['residual']=df_pf['targets']-df_pf['predictions']
```

```python
df_pf['difference%']=np.absolute(df_pf['residual']/df_pf['targets'])*100
```
